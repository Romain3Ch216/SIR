```python
def is_obstacle(image,seuil):
    if sum(sum(image))>seuil*image.shape[0]*image.shape[1]:
        return True
    else: return False
```

```python
def display(images):
    fig = plt.figure(figsize=(100, 100))
    columns = len(images)
    rows = 1
    for i in range(1, columns*rows +1):
        fig.add_subplot(rows, columns, i)
        plt.imshow(images[i-1])
    plt.show()
```

```python
frame_g_1 = cv2.imread('data/frame_g_25.jpg',0)
frame_d_1 = cv2.imread('data/frame_d_25.jpg',0)
stereo = cv2.StereoBM_create(numDisparities=64, blockSize=11)
disparity = stereo.compute(frame_g_1,frame_d_1)

images = [disparity,frame_g_1,frame_d_1]

#display(images)
```

```python
def home_made_convolution(image,size_filter):
    n,p = image.shape
    res = np.zeros((n-size_filter+1,p-size_filter+1))
    for i in range(n-size_filter+1):
        for j in range(p-size_filter+1):
            for k in range(-size_filter//2,size_filter//2):
                for m in range(-size_filter//2,size_filter//2):
                    res[i,j] += image[i+size_filter//2+k,j+size_filter//2+m]
    return res
```
