```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sc
```

```python
x = np.arange(0,100,0.1)
w = np.pi/10
y = [np.sin(w*e) for e in x]
```

```python
fig = plt.figure()

plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
c = []
for k in range(len(y)):
    L = [y[i+k]*y[i] for i in range(len(y)-k)]
    c.append(sum(L))
c = [e/max(c) for e in c]
z = [0]*(len(y)-1)
c = z + c
```

```python
ac = sc.correlate(y,y,mode="same")
ac = [e/max(ac) for e in ac]
xx = np.arange(0,200,0.1)[0:-1]
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

fig, ax = plt.subplots(figsize=(10, 8))

# Set axis ranges; by default this will put major ticks every 25.
ax.set_xlim(0, 200)
ax.set_ylim(-1.2, 1.2)

# Change major ticks to show every 20.
ax.xaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_major_locator(MultipleLocator(20))

# Change minor ticks to show every 5. (20/4 = 5)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

ax.grid(which='major', color=(0.1,0.1,0.1), linestyle='-')
ax.grid(which='minor', color=(0.2,0.2,0.2), linestyle=':')

plt.plot(x,ac)
plt.plot(x,y)
plt.plot(xx,c)
plt.show()
```
