```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import dateutil
```

```python
if 'Order Date' in csvFile.columns : 
    csvFile['Date'] = csvFile['Order Date'].dt.date
    csvFile['Time'] = csvFile['Order Date'].dt.time
    del csvFile['Order Date']

csvFile.head()

print('Nombre de commande par jour')
print(csvFile['Date'].value_counts().head())

```
